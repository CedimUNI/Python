##Literales Numéricos
### Enteros
Son aquellos números que no poseen parte decimal o fraccionaria
Se componen únicamente por caracteres numéricos y solo pueden incluir los caracteres **+**, **-** y **_**

> El caracter *underscore* (_) no altera el valor del número y solo sirve para una mejor lectura

Escribir `25` o `+25` representa un **número positivo** y `-25` representa un **número negativo**


````python
print(25)
print(+25)
print(-25)
````

Escribir `101` o `1_0_1` representa al mismo número en Python pero se nota más la utilidad del _underscore_ en casos de números más grandes como `1111111111` que se puede representar como `1_111_111_111` para una mejor legibilidad

````python
print(type(1_280_345))
````

### Binarios
Usar números en notación binaria en Python no es muy complicado, solo se debe agregar `0b` o `0B` al inicio del literal en sistema binario, por ejemplo `0b110111`, `0B11011010`, `0b1101`, ...


````python
print(0b110101011)
print(0B101)
````

###Octales
Para usar números en el sistema octal en Python se agrega `0o` o `0O` al inicio del literal, por ejemplo `0o1234567`, `0O7251012`, ... 

````python
print(0o777777)
print(0O10)
````

### Hexadecimales
Para usar números en el sistema hexadecimal se agrega `0x` o  `0X` al inicio del literal, por ejemplo `0xABCDEF`, `0X5123C`, ...

> `A,B,C,D,E,F` representan cifras del 10 al 15 respectivamente

Recordemos que esto debido a que estamos trabajando en el sistema de base 16. 

````python
print(0x1A)
print(0X10BC)
````

# Variables

Una caja en la cual se pueden guardar objetos y tiene un nombre conocido como identificador. En python es llamado un alias al objeto.
Veamos las siguientes operaciones:

````python
import math
a = int(input("Lado1: "))
b = int(input("Lado2: "))
c = int(input("lado3: "))
semiperimetro = (a+b+c)/2
area = math.sqrt((semiperimetro*(semiperimetro-a)*(semiperimetro-b)*(semiperimetro-c)))
print(area)
````
