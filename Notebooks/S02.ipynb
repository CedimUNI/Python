## Normas

### Norma general de un vector

Definimos la _norma-p_ de un vector de longitud $n$ como:

$||v||_p = \displaystyle\left[ \sum_{k=1}^{N}|v_k|^p\right]^{1/p}$

Las 4 normas más utilizadas son:
- $p=1$, la suma de los valores absoluto de los elementos.
- $p=2$, la magnitud del vector o longitud Euclidiana.
- $p=Inf$, máximo valor absoluto de un elemento.
- $p=-Inf$, mínimo valor absoluto de un elemento.

```python
v0 = np.array([-1, 1, -8, -10, 25, -5])
```

```python
np.linalg.norm(v0)
```

```python
np.linalg.norm(v0, ord=1)
```

```python
np.linalg.norm(v0, ord=np.inf)
```

```python
np.linalg.norm(v0, ord=-np.inf)
```

### Norma de una matriz

La norma de una matriz $X_{m,n}$ con $(m,n>=2)$ se aplica para los dos siguientes casos conocidos, la forma general queda como investigación para el estudiante:

```python
X = np.array([[1, -10, 0, 5],
              [-9, 3, -3, 8]], dtype='float64')
```

#### Máxima suma de valores absolutos de una columna
$||X||_1 = \max\limits_{1\leq j \leq n} \left( \displaystyle\sum_{i=1}^m|a_{ij}| \right)$

```python
np.linalg.norm(X, ord=1)
```

#### Máxima suma de valores absolutos de una fila

$||X||_∞ = \max\limits_{1\leq i \leq m} \left( \displaystyle\sum_{j=1}^m|a_{ij}| \right)$

```python
np.linalg.norm(X, ord=np.inf)
```

## Gráfica de Funciones de Transferencia

A lo largo de la carrera de ingeniería electrónica y mecatrónica, veremos un apartado importante, que es la ingeniería de control, lo fundamental cuando estemos estudiando esta área es el trabajar con las _funciones de transferencia_. Ello consiste en tener una relación entre la variable de salida y la variable de entrada en el dominio de Laplace.

Por ejemplo:

Sabemos que la ecuación diferencial que rige el Movimiento Armónico viene a ser:

<p align="center">
  <img src="https://www.monografias.com/trabajos96/contabilidad-analitica/image104.png">
</p>

$$
m\ddot{y}+b\dot{y}+ky = u
$$

Pasando al dominio de Laplace:
$$
ms^2Y(s) + bsY(s) + kY(s) = U(s)
$$

La función de transferencia tendría la siguiente forma:
$$
\frac{Y(s)}{U(s)} = \frac{1}{ms^2+bs+k}
$$

```python
from scipy import signal as sg
import numpy as np
import matplotlib.pyplot as plt

m = 5 # Masa en Kg
b = 1 # Coef de amortiguamiento (N*s/m)
k = 100 # Coef del resorte (N/m)
tf = 30 # Tiempo final

num = [1]
den = [m, b, k]

sys = sg.lti(num, den)
t = np.linspace(0, tf, 1000)
t, y = sg.step(sys, T=t)
plt.plot(t, y)
```

## Ploteo en 3D

Graficar:

$$
z = f(x,y) = sen(\sqrt{x^2+y^2})
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import ticker
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(subplot_kw = {'projection':'3d'})

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)

z = lambda x, y: np.sin(np.sqrt(x**2+y**2))

surf = ax.plot_surface(X, Y, z(X, Y), cmap = cm.viridis)

ax.set_zlim(-1.5, 1.5)
ax.zaxis.set_major_locator(LinearLocator(5))
ax.zaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.3f}'))
fig.colorbar(surf, shrink = 0.5, aspect = 10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)

z = lambda x, y: np.sin(np.sqrt(x**2+y**2))

surf = ax.plot_surface(X, Y, z(X, Y), cmap = cm.inferno)

ax.set_zlim(-1.5, 1.5)
fig.colorbar(surf, shrink = 0.5, aspect = 10)
plt.show()
```

### Coordenadas Polares

Graficar para $r \in [0, 1.25]$ y $\theta \in [0, 2\pi]$:

$$
f(r, \theta) = (r*cos(\theta), r*sen(\theta), (r^2-1)^2)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

r = np.linspace(0, 1.25, 500)
p = np.linspace(0, 2*np.pi, 500)
R, P = np.meshgrid(r, p)

x = lambda r, p: r*np.cos(p)
y = lambda r, p: r*np.sin(p)
z = lambda r, p: (r**2-1)**2

surf = ax.plot_surface(x(R, P), y(R, P), z(R, P), cmap = cm.plasma)

ax.set_zlim(0, 1)
ax.set_xlabel(r'$\phi_\mathrm{real}$')
ax.set_ylabel(r'$\phi_\mathrm{im}$')
ax.set_zlabel(r'$V(\phi)$')

fig.colorbar(surf, shrink=0.5, aspect=10)

plt.show()
```

## Subploteos

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure(figsize = (10, 4))

# Primer subploteo
ax = fig.add_subplot(1, 2, 1, projection = '3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y= np.meshgrid(x, y)
z = lambda x, y: np.sin(np.sqrt(x**2 + y**2))

surf = ax.plot_surface(X, Y, z(X, Y), cmap = cm.coolwarm)
ax.set_zlim(-1.5, 1.5)
fig.colorbar(surf, shrink=0.5, aspect=10)

# Segundo subploteo
ax = fig.add_subplot(122, projection='polar', aspect=0.5)

a = 2
theta = np.linspace(0, 2*np.pi, 100)
r = lambda theta: 2*a*np.sin(2*theta)

ax.plot(theta, r(theta))
ax.set_rmax(4)
ax.set_rticks(np.linspace(0,4,2))
ax.set_rlabel_position(-22.5)

fig.tight_layout()

plt.show()
```
